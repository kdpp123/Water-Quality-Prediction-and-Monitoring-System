FEATURE ENGINEERING

```python
# 1. Total Anions (Proxy for Total Dissolved Solids)
# This sums up the negative ions. High Anions usually means High Cations (Hardness).
combine_df['Total_Anions'] = combine_df['Cl (mg/L)'] + combine_df['SO4'] + combine_df['NO3'] + combine_df['HCO3'] + combine_df['CO3 (mg/L)']

# 2. Alkalinity (Carbonate Hardness proxy)
combine_df['Alkalinity'] = combine_df['HCO3'] + combine_df['CO3 (mg/L)']

# 3. EC to Chloride Ratio
# If this is High, it means there is conductivity NOT caused by Salt (likely Hardness).
# We add +1 to avoid dividing by zero errors.
combine_df['EC_Cl_Ratio'] = combine_df['EC (µS/cm at'] / (combine_df['Cl (mg/L)'] + 1)
```

```python
# 1. Define Constants
standards = {
    'pH': 8.5,
    'EC (µS/cm at': 500,
    'HCO3': 200,
    'Cl (mg/L)': 250,
    'F (mg/L)': 1.0,
    'SO4': 200,
    'NO3': 45,
    'Ca (mg/L)': 75,
    'Mg (mg/L)': 30,
    'Na (mg/L)': 200,
    'Total Hardness': 200
}

ideal = {
    'pH': 7.0,
    'EC (µS/cm at': 0,
    'HCO3': 0,
    'Cl (mg/L)': 0,
    'F (mg/L)': 0,
    'SO4': 0,
    'NO3': 0,
    'Ca (mg/L)': 0,
    'Mg (mg/L)': 0,
    'Na (mg/L)': 0,
    'Total Hardness': 0
}

# 2. Calculate Unit Weights (W_i)
# Formula: W_i = K / S_i  (Where K is proportionality constant)
# Your simplified method (1/S_i) is mathematically equivalent for the final ratio.
weights = {param: 1 / standards[param] for param in standards}

# 3. Compute WQI
# We start with 0 and add the components loop by loop
numerator = 0
denominator = 0

for param in standards:
    if param in combine_df.columns:
        # Get columns
        Vi = combine_df[param]             # Actual Value
        Vs = standards[param]      # Standard Value
        Vid = ideal[param]         # Ideal Value
        Wi = weights[param]        # Unit Weight

        # Quality Rating (Qi)
        # Note: abs() is sometimes used for pH, but your formula is standard direction
        Qi = ((Vi - Vid) / (Vs - Vid)) * 100

        # Add to sums
        numerator += (Qi * Wi)
        denominator += Wi
    else:
        print(f"Warning: '{param}' not found in dataset. Skipping.")

# Final Formula: WQI = Sum(Qi * Wi) / Sum(Wi)
combine_df["WQI"] = numerator / denominator

# 4. Assign Levels
def classify_wqi(wqi):
    if wqi <= 25: return "Excellent"
    elif wqi <= 50: return "Good"
    elif wqi <= 75: return "Poor"
    elif wqi <= 100: return "Very Poor"
    else: return "Unsafe"

combine_df["WQI_Level"] = combine_df["WQI"].apply(classify_wqi)
```

```python
import pandas as pd

# WHO/BIS Standards (S_i values)
standards = {
    'pH': 8.5,
    'EC (µS/cm at': 500,
    'HCO3': 200,
    'Cl (mg/L)': 250,
    'F (mg/L)': 1.0,
    'SO4': 200,
    'NO3': 45,
    'Ca (mg/L)': 75,
    'Mg (mg/L)': 30,
    'Na (mg/L)': 200,
    'Total Hardness': 200
}

# Ideal values for each parameter (V_ideal)
ideal = {
    'pH': 7.0,
    'EC (µS/cm at': 0,
    'HCO3': 0,
    'Cl (mg/L)': 0,
    'F (mg/L)': 0,
    'SO4': 0,
    'NO3': 0,
    'Ca (mg/L)': 0,
    'Mg (mg/L)': 0,
    'Na (mg/L)': 0,
    'Total Hardness': 0
}

# Calculate weights W_i = 1 / S_i
weights = {param: 1 / standards[param] for param in standards}

QiWi_list = []

# Compute Qi and Qi*Wi for each parameter
for param in standards:
    Vi = df[param]
    Vs = standards[param]
    Vid = ideal[param]
    Wi = weights[param]

    # Quality rating
    Qi = ((Vi - Vid) / (Vs - Vid)) * 100

    # Weighted Qi
    QiWi_list.append(Qi * Wi)

# Final WQI
df["WQI"] = sum(QiWi_list) / sum(weights.values())

# Assign WQI levels
def classify_wqi(wqi):
    if wqi <= 25:
        return "Excellent"
    elif wqi <= 50:
        return "Good"
    elif wqi <= 75:
        return "Poor"
    elif wqi <= 100:
        return "Very Poor"
    else:
        return "Unsafe"

df["WQI_Level"] = df["WQI"].apply(classify_wqi)

# Save output
df.to_excel("water_with_WQI.xlsx", index=False)
df.to_csv("water_with_WQI.csv", index=False)

print("WQI and WQI_Level columns added successfully!")
```

```python
feature = [
    "pH",
    "EC (µS/cm at",
    "CO3 (mg/L)",
    "HCO3",
    "Cl (mg/L)",
    "F (mg/L)",
    "SO4",
    "NO3",
    "Na (mg/L)",
    "K (mg/L)",
    "U (ppb)"
]

X_new = df_test[feature]

# Convert 'Total Hardness' to numeric and handle missing values
df_test['Total Hardness'] = pd.to_numeric(df_test['Total Hardness'], errors='coerce')
if df_test['Total Hardness'].isnull().any():
    df_test['Total Hardness'] = df_test['Total Hardness'].fillna(df_test['Total Hardness'].mean())

y_true = df_test["Total Hardness"]    # actual values from your CSV
```
